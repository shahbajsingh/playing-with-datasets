## plotting with iris dataset

```python
from sklearn import datasets
import pandas as pd

iris = datasets.load_iris()
```

```python
print("IRIS DATA")
```

```python
print("Type:\n", type(iris), "\n")

print("Keys:\n", iris.keys(), "\n")

print("Description:\n", iris.DESCR, "\n")

print("Feature names:\n", iris.feature_names, "\n")

print("Target:\n", iris.target, "\n")

print("Target names:\n", iris.target_names, "\n")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-whitegrid')
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['Species'] = iris.target
df
```

```python
print(df.info())
print(df.describe())
```

```python
features = iris.data.T
Labels = ['Setosa', 'Versicolor', 'Virginica']
colors = ['r', 'g', 'b']
classlbl = iris.target

fig = plt.figure(figsize = (12, 10))
ax = fig.add_subplot(111)

colors = np.asarray(colors)
colorslist = colors[classlbl]
Labels = np.asarray(Labels)
labellist = Labels[classlbl]

for x, y, s, c, l in zip(features[0], features[1], features[3], colorslist, labellist):
    ax.scatter(x, y, s = 100 * s, color = c, label = l, cmap = 'viridis', alpha = 0.2)
        
handles, labels = ax.get_legend_handles_labels()

newLeg = dict()

for h, l in zip(handles, labels):
    if l not in newLeg.keys():
        newLeg[l] = h
        
# create new handles and labels

handles = [ ]
labels = [ ]
for l in newLeg.keys():
    handles.append(newLeg[l])
    labels.append(l)
    
# create new legend

ax.legend(handles, labels, loc = 'lower right', frameon = True, prop = {'size': 20})
ax.set_xlabel(iris.feature_names[0], fontsize = '22')
ax.set_ylabel(iris.feature_names[1], fontsize = '22')
ax.set_title('Demonstration of Data Visualization with Iris', fontsize = '22')
plt.show()
```

## plotting with iris dataset in 3d

```python
from mpl_toolkits.mplot3d import Axes3D
features = iris.data.T
Labels = ['Setosa', 'Versicolor', 'Virginica']
colors = ['r', 'g', 'b']
classlbl = iris.target

fig = plt.figure(figsize = (12, 10))
ax = fig.add_subplot(111, projection = '3d')

colors = np.asarray(colors)
colorslist = colors[classlbl]
Labels = np.asarray(Labels)
labellist = Labels[classlbl]

for x, y, z, s, c, l in zip(features[0], features[1], features[2], features[3], colorslist, labellist):
    ax.scatter(x, y, z, s = 100 * s, color = c, label = l, cmap = 'viridis', alpha = 0.2)
    
handles, labels = ax.get_legend_handles_labels()

newLeg = dict()

for h, l in zip(handles, labels):
    if l not in newLeg.keys():
        newLeg[l] = h
        
# create new handles and labels

handles = [ ]
labels = [ ]

for l in newLeg.keys():
    handles.append(newLeg[l])
    labels.append(l)
    
# create new legend

ax.legend(handles, labels, loc = 'center left', frameon = True, prop = {'size': 10})
ax.set_xlabel(iris.feature_names[0], fontsize = '20')
ax.set_ylabel(iris.feature_names[1], fontsize = '20')
ax.set_zlabel(iris.feature_names[2], fontsize = '20')
ax.set_title('Demonstration of Data Visualization', fontsize = '22')
plt.show()
```
